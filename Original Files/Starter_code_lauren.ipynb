```python
import plotly.express as px

# Function to create a bar chart for a specific continent
def create_bar_chart_for_continent(df, continent):
    # Filter data for the given continent
    continent_df = df[df['Continent'] == continent]
    
    # Ensure we're working with years from 2000 to 2019
    continent_df = continent_df[continent_df['Year'].between(2000, 2019)]

    # Create a bar plot
    fig = px.bar(
        continent_df,
        x='Year',  # Set x-axis to Year
        y='Country', 
        color='Renewable energy share in the total final energy consumption (%)',  # Color by renewable energy percentage
        orientation='h',  # Horizontal bars
        title=f'Renewable Energy Share by Country for {continent} (2000-2019)',
        labels={'Renewable energy share in the total final energy consumption (%)': 'Renewable Energy (%)'},
        height=600,
        hover_data={'Year': True, 'Renewable energy share in the total final energy consumption (%)': True}  # Show year and renewable energy share on hover
    )
    
    # Customize layout
    fig.update_layout(
        xaxis_title='Year',
        yaxis_title='Country',
        coloraxis_colorbar=dict(title='Renewable Energy (%)'),  # Add title to the color scale
        xaxis=dict(tickmode='linear', tickvals=list(range(2000, 2020)))  # Set x-axis ticks for each year
    )
    
    # Show the figure
    fig.show()

# List of unique continents
continents = filtered_energy_table['Continent'].unique()

# Create bar charts for each continent
for continent in continents:
    create_bar_chart_for_continent(filtered_energy_table, continent)
```
